## Binary classification (C1W2L02)

Logistic regression is a statistical method used for binary classification.

Binary classification means that there are two possible classes for a given input. A good example is cat image classification, where the two classes are "cat" and "non cat".

It can be represented mathematically as follows:

$$
y \in \{0, 1 \}
$$

Where:

$y$ = the output label

$1$ = positive class (cat)

$0$ = negative class (non cat)

Matrices are often used to represent datasets, where matrix $X$ contains the features and $Y$ contains the labels.

Features are the input variables used to make predictions. In the case of cat image classification, the features are raw pixel values from which the model can learn patterns such as edges, eyes, ears, and whiskers.

$y$, as mentioned earlier, is the label indicating whether the image contains a cat or not. 

This can be represented as follows:

$$
(x^{(i)}, y^{(i)})
$$

In the video classification example, a 64x64 pixel image is used, this image has 3 color channels (RGB), so the total number of features is:

$$
n_x = 64 \times 64 \times 3 = 12288
$$

Where $n_x$ = number of features

$x$ is a real-valued vector, so $x \in \mathbb{R}^{n_x}$. In this case we have $x \in \mathbb{R}^{12288}$.

$y$ is the label and can be either 0 or 1, so $y \in \{0, 1\}$

$m$ = number of training examples = $(x^{(1)}, y^{(1)})$, $(x^{(2)}, y^{(2)})$, $...$ , $(x^{(m)}, y^{(m)})$

Example:

$$
X =
\begin{bmatrix}
\vert & \vert &        & \vert \\
x^{(1)} & x^{(2)} & \cdots & x^{(m)} \\
\vert & \vert &        & \vert
\end{bmatrix}
\in \mathbb{R}^{n_x \times m}
$$

$m$ = number of training examples (columns)

$n_x$ = number of features (rows)

$Y$ is represented as a row vector::

$$
Y = \begin{bmatrix} y^{(1)} & y^{(2)} & \cdots & y^{(m)} \end{bmatrix}
\in \mathbb{R}^{1 \times m} 
$$

In Python, this is translated as X.shape and Y.shape.
